#### Hospitals (num = 20)

```python
class HospitalGenerator:
    def __init__(self, seed = 0):
        self.fake = Faker('en_US')  # use en_US as language
        Faker.seed(seed)
        random.seed(seed)

        self.hospital_types = [
                'General Hospital',
                'Medical Center',
                'Hospital',
                'Regional Medical Center',
                'Community Hospital',
                'Memorial Hospital',
                'Health Center',
                'Medical Institute',
                'Clinic',
                'Healthcare System'
            ]

        # common naming way
        self.name_patterns = [
            'city',
            'saint',
            'university',
            'regional',
            'memorial',
            'community',
            'specialty'
        ]

        #
        self.specialties = [
            'Children\'s',
            'Women\'s',
            'Heart',
            'Cancer',
            'Orthopedic',
            'Eye',
            'Psychiatric',
            'Rehabilitation'
            ]

        self.city = None
        self.address = None


    def generate_hospital_address(self):
        """
        generate hospital address
        pattern: street name, city, state ZIP
        """
        address =  self.fake.address()
        # clean address
        address = address.replace('\n', ', ')
        patterns_to_remove = [
            r'\s+Apt\.?\s+\d+', # Apt xxx, Apt. xxx
            r'\s+Suite\s+\d+',             # Suite 123
            r'\s+Box\s+\d+',               # Box 123
        ]
        for pattern in patterns_to_remove:
            address = re.sub(pattern, '', address, flags=re.IGNORECASE)

        self.address = address.strip()

        # abstract city from address
        temp = address.split(',')
        if len(temp) >= 2:
            self.city = temp[-2].strip()
        else:  # otherwise create a new address
            self.generate_hospital_address()

    def generate_hospital_name(self):
        """
        generate hospital name in the different patterns
        """
        self.generate_hospital_address()
        city = self.city
        pattern = random.choice(self.name_patterns)

        if pattern == 'city':
            # pattern: [city name] + [type]
            hospital_type = random.choice(self.hospital_types)
            return f"{city} {hospital_type}"

        elif pattern == 'saint':
            # pattern: St. [name] + [type]
            first_name = self.fake.first_name()
            hospital_type = random.choice(self.hospital_types)

            # 50% add a " 's "
            if random.random() < 0.5:
                return f"St. {first_name}'s {hospital_type}"
            else:
                return f"St. {first_name} {hospital_type}"

        elif pattern == 'university':
            # pattern: [university name] + Medical Center/Hospital
            # use city name or random last name as name of university
            if random.random() < 0.5:
                university = city
            else:
                university = self.fake.last_name()

            suffixes = ['Medical Center', 'University Hospital',
                       'Teaching Hospital', 'Medical School Hospital']
            return f"{university} {random.choice(suffixes)}"

        elif pattern == 'regional':
            # pattern: [direction] + [region] + [type]
            directions = ['North', 'South', 'East', 'West',
                         'Northeast', 'Northwest', 'Southeast', 'Southwest']
            locations = ['Shore', 'Valley', 'Mountain', 'Bay',
                        'Lake', 'River', 'Hill', 'Coast']

            direction = random.choice(directions)
            location = random.choice(locations)
            hospital_type = random.choice(self.hospital_types)

            return f"{direction} {location} {hospital_type}"

        elif pattern == 'memorial':
            # pattern: [name] + Memorial Hospital
            name = random.choice([
                self.fake.last_name(),
                self.fake.first_name(),
                f"{self.fake.first_name()} {self.fake.last_name()}"
            ])
            return f"{name} Memorial Hospital"

        elif pattern == 'community':
            # pattern: [community name] + Community Hospital
            return f"{city} Community Hospital"

        elif pattern == 'specialty':
            # pattern: [city name] + [specialty] + Hospital
            specialty = random.choice(self.specialties)
            return f"{city} {specialty} Hospital"

        else:
            # default
            return f"{city} Medical Center"

    def generate(self, n):
        """
        generate n hospital data
        """
        res = []
        visited = set()
        i = 1

        while i <= n:
            name = self.generate_hospital_name()
            if name in visited or self.address is None:
                continue

            visited.add(name)
            res.append({
                'hospital_id': i,
                'hospital_name': name,
                'hospital_address': self.address
            })

            i += 1

        return pd.DataFrame(res)
```

```python
class DepartmentGenerator:
    def __init__(self, departments, df_hospital_path = 'hospitals.csv', seed = 0):
        self.departments = departments

        file_path = Path(df_hospital_path)
        if file_path.exists():
            self.df_hospital = pd.read_csv(file_path)
        elif Path('hospitals.csv').exists():
            self.df_hospital = pd.read_csv('hospitals.csv')
        else:
            raise ValueError('Hospitals table not exists.')

        self.n_hospital = self.df_hospital.shape[0]
        self.specialties = [
            'Children\'s',
            'Women\'s',
            'Heart',
            'Cancer',
            'Orthopedic',
            'Eye',
            'Psychiatric',
            'Rehabilitation'
            ]
        self.specialties_map_department = {
            'Children\'s': 'Pediatrics',
            'Women\'s': 'Obstetrics and Gynecology',
            'Heart': "Cardiology",
            'Cancer': "Oncology",
            'Orthopedic': "Orthopedics",
            'Eye': "Ophthalmology",
            'Psychiatric': "Psychiatry",
            'Rehabilitation': "Physical Medicine and Rehabilitation"
        }
        random.seed(seed)


    def generate(self):
        """
        generate departments for all hospitals
        """
        res = []
        id = 1
        for i in range(self.n_hospital):
            hospital_name = self.df_hospital.iloc[i, 2]
            department_list = set()
            # if specialty hospital
            for specialty in self.specialties:
                if specialty in hospital_name:
                    department_list.add(self.specialties_map_department[specialty])

            # randomly choose m departments
            m = random.randint(1, len(self.departments))
            temp = random.sample(self.departments, m)
            department_list.update(temp)

            for department in department_list:
                res.append({
                    'department_id': id,
                    'department_name': department,
                    'hospital_id': i + 1
                })
                id += 1


        return pd.DataFrame(res)
```

```python
class DoctorGenerator:
    def __init__(self, specialties, df_department_path = 'departments.csv', df_hospital_path = 'hospitals.csv', seed=0):
        self.fake = Faker()
        self.seed = seed
        Faker.seed(seed)
        random.seed(seed)

        self.specialties = specialties

        file_path = Path(df_department_path)
        if file_path.exists():
            self.df_department = pd.read_csv(file_path)
        elif Path('departments.csv').exists():
            self.df_department = pd.read_csv('departments.csv')
        else:
            raise ValueError('Departments table not exists.')

        file_path = Path(df_hospital_path)
        if file_path.exists():
            self.df_hospitals = pd.read_csv(file_path)
        elif Path('hospitals.csv').exists():
            self.df_hospitals = pd.read_csv('hospitals.csv')
        else:
            raise ValueError('Hospitals table not exists.')


    def generate_email(self, first_name, last_name, hospital_id):
        """
        generate email in standardized format for each hospital
        """

        domain = self.df_hospitals[self.df_hospitals['hospital_id'] == hospital_id]['hospital_name'].values
        domain = domain[0].split(' ')

        abbreviation = ''.join([word[0] for word in domain])

        email = f"{first_name.lower()}.{last_name.lower()}@{abbreviation}.com"

        return email

    def generate_doctors_for_hospital(self, hospital_id, n_doctors=50):
        """
        generate doctor for all departments
        """
        hospital_departments = self.df_department[
            self.df_department['hospital_id'] == hospital_id
        ]

        if len(hospital_departments) == 0:
            return []

        if n_doctors < len(hospital_departments):
            print(f'n_doctors is too small for hospital {hospital_id}, increase n_doctors to 2 * number of departments.')
            n_doctors = 2 * len(hospital_departments)

        doctors = []

        random.seed(self.seed)
        # 1. make sure that all departments are assigned at least one doctor
        for _, dept_row in hospital_departments.iterrows():
            department_name = dept_row['department_name']
            department_id = dept_row['department_id']

            # randomly select a specialty
            if department_name in self.specialties:
                specialization = random.choice(self.specialties[department_name])
            else:
                specialization = department_name  # directly use department name

            first_name = self.fake.first_name()
            last_name = self.fake.last_name()

            doctor = {
                'first_name': first_name,
                'last_name': last_name,
                'specialization': specialization,
                'phone_number': self.fake.phone_number(),
                'years_experience': random.randint(1, 35),
                'hospital_id': hospital_id,
                'email': self.generate_email(first_name, last_name, hospital_id),
                'department_id': department_id,
                'doctor_rating': np.round(min(max(np.random.normal(loc=3, scale=0.5), 1), 5), 2)
            }
            doctors.append(doctor)

        # 2. randomly assign the rest of the doctors to departments excluding the doctors above
        remaining_doctors = n_doctors - len(hospital_departments)

        for _ in range(remaining_doctors):
            # randomly select a department
            dept_row = hospital_departments.sample(n=1).iloc[0]
            department_name = dept_row['department_name']
            department_id = dept_row['department_id']

            # randomly select a specialty
            if department_name in self.specialties:
                specialization = random.choice(self.specialties[department_name])
            else:
                specialization = department_name

            first_name = self.fake.first_name()
            last_name = self.fake.last_name()

            doctor = {
                'first_name': first_name,
                'last_name': last_name,
                'specialization': specialization,
                'phone_number': self.fake.phone_number(),
                'years_experience': random.randint(1, 35),
                'hospital_id': hospital_id,
                'email': self.generate_email(first_name, last_name, hospital_id),
                'department_id': department_id,
                'doctor_rating': np.round(min(max(np.random.normal(loc=3.5, scale=0.5), 1), 5), 2)
            }
            doctors.append(doctor)

        return doctors

    def generate(self, n_doctors_per_hospital=50):
        """
        generate doctors for all hospitals
        """
        all_doctors = []
        doctor_id = 1

        # get all hospital_id
        hospital_ids = self.df_department['hospital_id'].unique()

        for hospital_id in hospital_ids:
            doctors = self.generate_doctors_for_hospital(hospital_id, n_doctors_per_hospital)

            # add doctor_id
            for doctor in doctors:
                doctor['doctor_id'] = doctor_id
                all_doctors.append(doctor)
                doctor_id += 1

        df = pd.DataFrame(all_doctors)

        return df[['doctor_id', 'first_name', 'last_name', 'specialization', 'phone_number', 'years_experience',
                'hospital_id', 'email', 'department_id', 'doctor_rating']]
```

```python
class PatientGenerator:
    def __init__(self, seed=0):
        self.fake = Faker()
        Faker.seed(seed)
        random.seed(seed)

        # insurance providers
        self.insurance_providers = [
            'Blue Cross Blue Shield',
            'UnitedHealthcare',
            'Aetna',
            'Cigna',
            'Humana',
            'Kaiser Permanente',
            'Anthem',
            'Medicare',
            'Medicaid',
            None
        ]

    def generate_date_of_birth(self):
        """
        generate date of birth (age between 0 and 100)
        """
        today = dt.date.today()
        age = random.randint(0, 100)
        birth_year = today.year - age
        start_date = dt.date(birth_year, 1, 1)
        end_date = dt.date(birth_year, 12, 31)

        # ensuring no dates are randomly generated from the future
        if end_date > today:
            end_date = today

        return self.fake.date_between(start_date=start_date, end_date=end_date)

    def generate_registration_date(self, dob):
        """
        generate registration date with 80% of patients registered within past 5 years
        """
        today = dt.date.today()

        if random.random() < 0.8:
            earliest_date = max(dob, today - timedelta(days=5*365))
        else:
            earliest_date = dob

        return self.fake.date_between(start_date=earliest_date, end_date=today)


    def generate_insurance_number(self, insurance_provider):
        """
        generate insurance number if insurance provider exists else return None
        """
        if insurance_provider is None:
            return None

        # format: letters*3 + numbers*9
        prefix = ''.join(random.choices('ABCDEFGHIJKLMNOPQRSTUVWXYZ', k=3))
        number = ''.join(random.choices('0123456789', k=9))
        return f"{prefix}{number}"

    def generate_email(self, first_name, last_name):
        """
        generate emails for 70% patients
        """
        if random.random() < 0.3:
            return None

        domains = ['gmail.com', 'yahoo.com', 'hotmail.com', 'outlook.com', 'icloud.com']
        # add random numbers
        random_num = random.randint(1, 999)
        email = f"{first_name.lower()}.{last_name.lower()}{random_num}@{random.choice(domains)}"
        return email


    def generate(self, n):
        """
        generate patient data for n number of patients
        """
        patients = []

        for i in range(n):

            gender = random.choice(['M', 'F'])
            # generating realistic first names that are gender-appropriate
            if gender == 'M':
                firstName = self.fake.first_name_male()
            else:
                firstName = self.fake.first_name_female()
            lastName = self.fake.last_name()
            address = self.fake.address()
            address = address.replace('\n', ', ').strip()

            dob = self.generate_date_of_birth()
            registration_date = self.generate_registration_date(dob)

            insurance_provider = random.choice(self.insurance_providers)
            insurance_number = self.generate_insurance_number(insurance_provider)

            patient = {
                # creating a patient with the following characteristics
                'Patient ID': i + 1,
                'First Name': firstName,
                'Last Name': lastName,
                'Gender': gender,
                'DOB': dob,
                'Phone Number': self.fake.phone_number(),
                'Address': address,
                'Registration Date': registration_date,
                'Insurance Provider': insurance_provider,
                'Insurance Number': insurance_number,
                'Email': self.generate_email(firstName, lastName),
                'SSN': self.fake.ssn()
            }

            patients.append(patient)


        df = pd.DataFrame(patients)

        # assigning convenint column order for the generated .csv file/data frame
        column_order = [
            'Patient ID', 'First Name', 'Last Name', 'Gender',
            'DOB', 'Phone Number', 'Address', 'Registration Date',
            'Insurance Provider', 'Insurance Number', 'Email', 'SSN'
        ]

        return df[column_order]
```

```python
class AppointmentGenerator:
    def __init__(self, reasons_for_visit, reason_to_departments, reason_to_appointment_type, df_doctor_path = 'doctors.csv', df_department_path = 'departments.csv',
    df_patient_path = 'patients.csv', start_date = None, end_date = dt.date.today(), seed = 0):
        self.fake = Faker()
        Faker.seed(seed)
        random.seed(seed)

        if start_date is None:
            self.start_date = end_date - dt.timedelta(days=5*365)  # record appointments for past 5 years
        else:
            self.start_date = start_date
        self.end_date = end_date + dt.timedelta(days=365)  # accept appointments for 1 year in the future

        self.reasons_for_visit = reasons_for_visit
        self.reason_to_departments = reason_to_departments
        self.reason_to_appointment_type = reason_to_appointment_type
        self.status = ['Scheduled', 'Completed', 'Cancelled']

        file_path_dept = Path(df_department_path)
        if file_path_dept.exists():
            self.df_departments = pd.read_csv(file_path_dept)
        elif Path('departments.csv').exists():
            self.df_departments = pd.read_csv('departments.csv')
        else:
            raise ValueError('Departments table not exists.')

        file_path_doc = Path(df_doctor_path)
        if file_path_doc.exists():
            self.df_doctors = pd.read_csv(file_path_doc)
        elif Path('doctors.csv').exists():
            self.df_doctors = pd.read_csv('doctors.csv')
        else:
            raise ValueError('Doctors table not exists.')

        file_path_pat = Path(df_patient_path)
        if file_path_pat.exists():
            self.df_patients = pd.read_csv(file_path_pat)
        elif Path('patients.csv').exists():
            self.df_patients = pd.read_csv('patients.csv')
        else:
            raise ValueError('Patients table not exists.')

        self.patient_appointment_history = {}  # Track appointments per patient

    def generate_appointment(self, patient_id=None, is_followup=False, previous_appointment=None):
        """
        generate appointment for a patient
        """
        # select patient (or use provided patient_id for follow-ups)
        if patient_id:
            patient = self.df_patients[self.df_patients['Patient ID'] == patient_id].iloc[0]
        else:
            patient = self.df_patients.sample(n=1).iloc[0]

        patient_id = patient['Patient ID']

        # determine if this should be a follow-up based on patient history
        if not is_followup and patient_id in self.patient_appointment_history:
            completed_appointments = [
                apt for apt in self.patient_appointment_history[patient_id]
                if apt['status'] == 'Completed'
            ]
            # 30% chance of follow-up if patient has completed appointments
            if completed_appointments and random.random() < 0.3:
                is_followup = True
                previous_appointment = random.choice(completed_appointments)

        # select reason and type
        if is_followup and previous_appointment:
            # follow-up uses same reason as previous appointment
            reason = previous_appointment['reason_for_visit']
            appointment_type = random.choice(['Regular', 'Diagnostic', 'Treatment'])
            occurance = 'Follow-up'

            # appointment date should be after previous appointment
            prev_date = previous_appointment['appointment_date']
            if isinstance(prev_date, str):
                prev_date = dt.datetime.strptime(prev_date, '%Y-%m-%d').date()

            appointment_start = prev_date + dt.timedelta(days=7)  # at least 1 week later
            appointment_start = max(appointment_start, self.start_date)
        else:
            reason = random.choice(self.reasons_for_visit)
            appointment_type = self.reason_to_appointment_type[reason]
            while appointment_type == "Treatment":
                reason = random.choice(self.reasons_for_visit)
                appointment_type = self.reason_to_appointment_type[reason]
            occurance = 'First-time'

            patient_reg_date = pd.to_datetime(patient['Registration Date']).date()
            appointment_start = max(patient_reg_date, self.start_date)

        available_departments = self.reason_to_departments[reason]
        department = random.choice(available_departments)

        department_records = self.df_departments[self.df_departments['department_name'] == department]
        if department_records.empty:
            return None

        selected_dept = department_records.sample(n=1).iloc[0]
        department_id = selected_dept['department_id']

        available_doctors = self.df_doctors[self.df_doctors['department_id'] == department_id]
        if len(available_doctors) == 0:
            return None

        doctor = available_doctors.sample(n=1).iloc[0]

        # generate appointment date and time
        appointment_date = self.fake.date_between(
            start_date=appointment_start,
            end_date=self.end_date
        )

        hour = random.randint(8, 16)
        minute = random.choice([0, 15, 30, 45])
        appointment_time = dt.time(hour, minute)

        # generate status
        if appointment_date > dt.date.today():
            status = 'Scheduled'
        else:
            if appointment_type in ['Treatment', 'Others']:
                status = 'Cancelled' if random.random() < 0.1 else 'Completed'
            else:
                status = random.choices(['Cancelled', 'Completed'], weights=[2, 8], k=1)[0]

        appointment = {
            'patient_id': patient_id,
            'doctor_id': doctor['doctor_id'],
            'reason_for_visit': reason,
            'appointment_type': appointment_type,
            'appointment_occurance': occurance,
            'status': status,
            'appointment_time': appointment_time,
            'appointment_date': appointment_date
        }

        return appointment

    def generate(self, n):
        """
        generate appointments for n number of patients
        """
        res = []

        for i in range(n):
            temp = self.generate_appointment()
            if temp is None:
                continue

            temp['appointment_id'] = i + 1
            res.append(temp)

            # track appointment history per patient
            patient_id = temp['patient_id']
            if patient_id not in self.patient_appointment_history:
                self.patient_appointment_history[patient_id] = []
            self.patient_appointment_history[patient_id].append(temp)

        # sort by patient_id and date to maintain chronological order
        df = pd.DataFrame(res)
        #df = df.sort_values(['patient_id', 'appointment_date']).reset_index(drop=True)

        columns = ['appointment_id', 'patient_id', 'doctor_id', 'reason_for_visit',
                   'appointment_type', 'appointment_occurance', 'status',
                   'appointment_time', 'appointment_date']
        return df[columns]
```

```python
class TreatmentGenerator:
    def __init__(self, reason_to_treatment_types, treatment_cost_ranges, treatment_type_to_description_level,
     df_appointments_path = 'appointments.csv', seed = 0):
        self.faker = Faker()
        Faker.seed(seed)
        random.seed(seed)
        self.reason_to_treatment_types = reason_to_treatment_types
        self.treatment_cost_ranges = treatment_cost_ranges
        self.treatment_type_to_description_level = treatment_type_to_description_level

        file_path = Path(df_appointments_path)
        if file_path.exists():
            self.df_appointments = pd.read_csv(file_path)
        elif Path('appointments.csv').exists():
            self.df_appointments = pd.read_csv('appointments.csv')
        else:
            raise ValueError('Appointments table not exists.')

    def generate(self, end_date = dt.date.today()):
        """
        generate treatment for every past completed appointment
        """
        res = []
        treatment_id = 1
        for _, appointment in self.df_appointments.iterrows():
            if appointment['status'] == 'Completed':

                # get appointment id
                appointment_id = appointment['appointment_id']

                # get appointment date
                start_date = pd.Timestamp(appointment['appointment_date']).date()  # first transfer to pd.timestamp, then to pydatetime

                # get reason for visit from appointments
                reason = appointment['reason_for_visit']

                # get all possible treatments
                available_treatments = self.reason_to_treatment_types.get(reason)

                # randomly select treatments
                treatment_types = random.sample(available_treatments, k = min(random.randint(1, 3), len(available_treatments)))

                # get cost
                total_cost = 0
                for treatment in treatment_types:
                    cost_range = self.treatment_cost_ranges.get(treatment, (100, 1000))
                    total_cost += random.uniform(cost_range[0], cost_range[1])
                total_cost = round(total_cost, 2)

                # get descroption
                descriptions = [treatment_type_to_description_level[treatment] for treatment in treatment_types]
                if 'Advanced protocol' in descriptions:
                    description = 'Advanced protocol'
                elif 'Standard procedure' in descriptions:
                    description = 'Standard procedure'
                else:
                    description = 'Basic screening'

                # get treatment date
                if random.random() < 0.1:
                    # treatment may be 1-7 days later
                    days_after = random.randint(1, 7)
                    treatment_date = start_date + dt.timedelta(days=days_after)
                    # make sure it's not later than end_date(today)
                    if treatment_date > end_date:
                        treatment_date = end_date
                else:
                    treatment_date = start_date

                res.append({
                    'treatment_id': treatment_id,
                    'appointment_id': appointment_id,
                    'treatment_types': ', '.join(treatment_types),
                    'description': description,
                    'cost': total_cost,
                    'treatment_date': treatment_date,
                })

                treatment_id += 1

        df = pd.DataFrame(res)

        return df
```

```python
class BillGenerator:
    def __init__(self, df_treatments_path = 'treatments.csv', seed = 0):
        random.seed(seed)

        file_path = Path(df_treatments_path)
        if file_path.exists():
            self.df_treatments = pd.read_csv(file_path)
        elif Path('treatments.csv').exists():
            self.df_treatments = pd.read_csv('treatments.csv')
        else:
            raise ValueError('Treatments table not exists.')

        self.methods = ['Cash', 'Card', 'Insurance']

    def generate(self, end_date = dt.date.today()):
        """
        generate bill for every treatment
        """
        res = []
        id = 1
        for _, treatment in self.df_treatments.iterrows():
            treatment_id = treatment['treatment_id']
            appointment_id = treatment['appointment_id']
            cost = treatment['cost']
            bill_date = pd.Timestamp(treatment['treatment_date']).date()  # assume that the bills are usually paid at the treatment day

            method = random.choice(self.methods)
            if random.random() < 0.1:  # randomly generate some failed transaction
                status = 'Failed'
            else:
                if bill_date > end_date - dt.timedelta(days=2):
                    status = 'Pending'
                else:
                    status = 'Paid'

            res.append({
                'bill_id': id,
                'treatment_id': treatment_id,
                'bill_date': bill_date,
                'amount': cost,
                'payment_method': method,
                'payment_status': status,
                })

            id += 1

        return pd.DataFrame(res)
```

```python
# functions to generate lab results

def generate_patient_labs(patientList, maxVisitNum=10, seed=23):
    """
    Generating lab results for patients with multiple visits over time. Usually multiple visits
    are diagnostic appointments in which lab results may be taken for a patient.

    Parameters:
        * patientList: dataframe holding all patient identifier data
        (in our case patients.csv) (it should also have patient_id, registration_date)
        * maxVisitNum: The maximum number of lab visits per patient - default is a max of 10
        * seed: The seed ensures the data generated is the same random data each time
    """
    random.seed(seed)

    labChoice = [
        "Blood Chemistry Panel",
        "Complete Blood Count",
        "Hormones",
        "Total Lipid Panel",
        "Urinalysis",
        "Vitamin D",
        "Thyroid Panel",
        "HbA1c",
        "Iron Studies",
        "Inflammatory Markers",
        "Pregnancy Test"
    ]

    labs = []

    for _, patient in patientList.iterrows():

        # determining the number of visits for each patient in patientList
        visits = random.randint(1, maxVisitNum)

        # identify the patient's registration date
        registrationDate = pd.to_datetime(patient["Registration Date"])

        # calculating the number of days the patient has been registered into the database
        days_since_reg = (datetime.today() - registrationDate).days
        if days_since_reg < 1:
            days_since_reg = 1

        # generating random visit dates since the day the patient registered at the hospital
        visit_offsets = sorted(random.sample(range(0, days_since_reg + 1), min(visits, days_since_reg + 1)))

        for v, offset in enumerate(visit_offsets, 1):
            visitDate = registrationDate + timedelta(days=offset)
            lab = random.choice(labChoice)

            # generating test results based on the type of lab
            testResults = generate_test_results(lab, patient["Gender"])

            record = {
                "Patient ID": patient["Patient ID"],
                "First Name": patient["First Name"],
                "Last Name": patient["Last Name"],
                "Gender": patient["Gender"],
                "DOB": patient["DOB"],
                "Date of Visit": visitDate.strftime("%Y-%m-%d"),
                "Lab Type": lab,
                **testResults # flattening test results into separate columns
            }
            labs.append(record)

    return pd.DataFrame(labs)


def generate_test_results(lab, gender):
    """
    Generates test results based on the type of lab and gender. Most labs have
    different parameter ranges for men and women.
    """

    if lab == "Blood Chemistry Panel":
        return {
            # found realistic ranges for each parameter
            "Blood Glucose (mg/dL)": round(random.uniform(70, 110), 1),
            "Sodium (mmol/L)": round(random.uniform(135, 145), 1),
            "Potassium (mmol/L)": round(random.uniform(3.5, 5.0), 2),
            "Chloride (mmol/L)": round(random.uniform(98, 107), 1),
            "Bicarbonate (HCO3-) (mmol/L)": round(random.uniform(22, 29), 1),
            "Blood Urea Nitrogen (BUN) (mg/dL)": round(random.uniform(7, 20), 1),
            "Creatinine (mg/dL)": round(random.uniform(0.6, 1.3), 2),
            "Calcium (mg/dL)": round(random.uniform(8.5, 10.5), 1),
            "Liver Enzymes (ALT, U/L)": round(random.uniform(7, 56), 1),
            "Bilirubin (mg/dL)": round(random.uniform(0.1, 1.2), 1),
            "Albumin (g/dL)": round(random.uniform(3.5, 5.0), 1)
        }

    elif lab == "Complete Blood Count":
        if gender == "M":
            hemoglobin = round(random.uniform(13.5, 17.5), 1)
            hematocrit = round(random.uniform(41, 53), 1)
        else:
            hemoglobin = round(random.uniform(12.0, 15.5), 1)
            hematocrit = round(random.uniform(36, 46), 1)
        return {
            "Hemoglobin (g/dL)": hemoglobin,
            "Hematocrit (%)": hematocrit,
            "WBC (10^3/L)": round(random.uniform(4.0, 11.0), 1),
            "RBC (10^6/L)": round(random.uniform(4.2, 5.9), 1),
            "Platelets (10^3/L)": round(random.uniform(150, 450), 0)
        }

    elif lab == "Hormones":
      # assigning new ranges for some hormones that depend on gender
        if gender == "M":
          progesterone = round(random.uniform(0.1, 1.2), 1)
          estrogen = round(random.uniform(10, 40), 1)
          testosterone = round(random.uniform(300, 1000), 1)
        else:
          progesterone = round(random.uniform(0.1, 25), 1)
          estrogen = round(random.uniform(30, 350), 1)
          testosterone = round(random.uniform(15, 75), 1)
        return {
            "TSH (mIU/L)": round(random.uniform(0.5, 4.5), 2),
            "T3 (nmol/L)": round(random.uniform(1.0, 2.5), 2),
            "T4 (g/dL)": round(random.uniform(4.5, 12.0), 1),
            "Cortisol (g/dL)": round(random.uniform(5, 25), 1),
            "Insulin (U/mL)": round(random.uniform(2, 25), 1),
            "Estrogen (pg/mL)": estrogen,
            "Progesterone (ng/mL)": progesterone,
            "Testosterone (ng/dL)": testosterone
        }

    elif lab == "Total Lipid Panel":
        return {
            "Total Cholesterol (mg/dL)": round(random.uniform(125, 200), 1),
            "LDL (mg/dL)": round(random.uniform(50, 130), 1),
            "HDL (mg/dL)": round(random.uniform(40, 60), 1),
            "Triglycerides (mg/dL)": round(random.uniform(50, 200), 1)
        }

    elif lab == "Urinalysis":
        return {
            "eGFR (mL/min/1.73m)": round(random.uniform(60, 120), 1),
            "Urine Protein (g/dL)": round(random.uniform(0, 0.3), 2),
            "Urine Glucose (mg/dL)": round(random.uniform(0, 15), 1),
            "Ketones (mg/dL)": round(random.uniform(0, 5), 1),
            "RBCs (x10^6/L)": round(random.uniform(0, 5), 1),
            "pH": round(random.uniform(4.5, 8.0), 1),
            "Nitrite (mg/dL)": round(random.uniform(0, 1), 2)
        }

    elif lab == "Vitamin D":
        d2 = round(random.uniform(0, 30), 1)
        d3 = round(random.uniform(10, 50), 1)
        return {
            # performing calculations for total Vitamin D count
            "Vitamin D2 (ng/mL)": d2,
            "Vitamin D3 (ng/mL)": d3,
            "Vitamin D Total (ng/mL)": d2+d3
        }

    elif lab == "Thyroid Panel":
        return {
            "TSH (mIU/L)": round(random.uniform(0.5, 4.5), 2),
            "Free T4 (ng/dL)": round(random.uniform(0.8, 1.8), 2),
            "Free T3 (pg/mL)": round(random.uniform(2.0, 4.4), 2)
        }

    elif lab == "HbA1c":
        return {
            "HbA1c (%)": round(random.uniform(4.5, 6.5), 2)
        }

    elif lab == "Iron Studies":
        return {
            "Serum Iron (g/dL)": round(random.uniform(40, 160), 1),
            "Ferritin (ng/mL)": round(random.uniform(12, 300), 1),
            "TIBC (g/dL)": round(random.uniform(240, 450), 1),
            "Transferrin Saturation (%)": round(random.uniform(15, 50), 1)
        }

    elif lab == "Inflammatory Markers":
        return {
            "CRP (mg/L)": round(random.uniform(0.1, 10.0), 2),
            "ESR (mm/hr)": round(random.uniform(0, 30), 1)
        }

    elif lab == "Pregnancy Test":
      # Human Chorionic Gonadotropin number tracks pregnancy - the higher the #, the further along the pregnancy
        hCGquant = round(random.uniform(0, 250000), 1)
        if hCGquant < 5:
          pregnancy = 'Not Pregnant'
        elif hCGquant < 50 and hCGquant >= 5:
          pregnancy = '3 Weeks Pregnant'
        elif hCGquant < 430 and hCGquant >= 50:
          pregnancy = '4 Weeks Pregnant'
        elif hCGquant < 229000 and hCGquant >= 430:
          pregnancy = '7-8 Weeks Pregnant'
        elif hCGquant < 288000 and hCGquant >= 229000:
          pregnancy = '10-12 Weeks Pregnant'
        return {
            "hCG Quantitative (mIU/mL)": hCGquant,
            "Pregnancy Prediction": pregnancy
        }

    else:
        return {"Unknown Lab": "Unknown Lab"}

    return {}
```

#### Hospitals

```python
# 20 hospitals

# CREATE TABLE Hospitals (
#     hospital_id INT PRIMARY KEY AUTO_INCREMENT,
#     hospital_name VARCHAR(45),
#     hospital_address TEXT
# );

import re

class HospitalGenerator:
    def __init__(self, seed = 0):
        self.fake = Faker('en_US')  # use en_US as language
        Faker.seed(seed)
        random.seed(seed)

        self.hospital_types = [
                'General Hospital',
                'Medical Center',
                'Hospital',
                'Regional Medical Center',
                'Community Hospital',
                'Memorial Hospital',
                'Health Center',
                'Medical Institute',
                'Clinic',
                'Healthcare System'
            ]
            
        # common naming way
        self.name_patterns = [
            'city',
            'saint',
            'university',
            'regional',
            'memorial',
            'community',
            'specialty'
        ]
        
        # 
        self.specialties = [
            'Children\'s',
            'Women\'s',
            'Heart',
            'Cancer',
            'Orthopedic',
            'Eye',
            'Psychiatric',
            'Rehabilitation'
            ]

        self.city = None
        self.address = None


    def generate_hospital_address(self):
        """
        pattern: street name, city, state ZIP
        """
        address =  self.fake.address()
        address = address.replace('\n', ', ')
        patterns_to_remove = [
            r'\s+Apt\.?\s+\d+', # Apt xxx, Apt. xxx
            r'\s+Suite\s+\d+',             # Suite 123
            r'\s+Box\s+\d+',               # Box 123
        ]
        for pattern in patterns_to_remove:
            address = re.sub(pattern, '', address, flags=re.IGNORECASE)

        self.address = address.strip()

        # abstract city from address
        temp = address.split(',')
        if len(temp) >= 2:
            self.city = temp[-2].strip()
        else:  # otherwise create a new address
            self.generate_hospital_address()
    
    def generate_hospital_name(self):
        self.generate_hospital_address()
        city = self.city
        pattern = random.choice(self.name_patterns)

        if pattern == 'city':
            # pattern: [city name] + [type]
            # e.g. "Boston General Hospital"
            hospital_type = random.choice(self.hospital_types)
            return f"{city} {hospital_type}"
        
        elif pattern == 'saint':
            # pattern: St. [name] + [type]
            # e.g. "St. Mary's Medical Center"
            first_name = self.fake.first_name()
            hospital_type = random.choice(self.hospital_types)
            
            # 50% add a " 's " 
            if random.random() < 0.5:
                return f"St. {first_name}'s {hospital_type}"
            else:
                return f"St. {first_name} {hospital_type}"
        
        elif pattern == 'university':
            # pattern: [university name] + Medical Center/Hospital
            # e.g. "Harvard Medical Center"
            # use city name or random last name as name of university
            if random.random() < 0.5:
                university = city
            else:
                university = self.fake.last_name()
            
            suffixes = ['Medical Center', 'University Hospital', 
                       'Teaching Hospital', 'Medical School Hospital']
            return f"{university} {random.choice(suffixes)}"
        
        elif pattern == 'regional':
            # pattern: [direction] + [region] + [type]
            # e.g. "North Shore Regional Medical Center"
            directions = ['North', 'South', 'East', 'West', 
                         'Northeast', 'Northwest', 'Southeast', 'Southwest']
            locations = ['Shore', 'Valley', 'Mountain', 'Bay', 
                        'Lake', 'River', 'Hill', 'Coast']
            
            direction = random.choice(directions)
            location = random.choice(locations)
            hospital_type = random.choice(self.hospital_types)
            
            return f"{direction} {location} {hospital_type}"
        
        elif pattern == 'memorial':
            # pattern: [name] + Memorial Hospital
            # e.g. "Johnson Memorial Hospital"
            name = random.choice([
                self.fake.last_name(),
                self.fake.first_name(),
                f"{self.fake.first_name()} {self.fake.last_name()}"
            ])
            return f"{name} Memorial Hospital"
        
        elif pattern == 'community':
            # pattern: [community name] + Community Hospital
            # e.g. "Springfield Community Hospital"
            return f"{city} Community Hospital"
        
        elif pattern == 'specialty':
            # pattern: [city name] + [specialty] + Hospital
            # e.g. "Boston Children's Hospital"
            specialty = random.choice(self.specialties)
            return f"{city} {specialty} Hospital"
        
        else:
            # default
            return f"{city} Medical Center"
    
    def generate(self, n):
        """
        generate n hospital data
        """
        res = []
        visited = set()
        i = 1

        while i <= n:
            name = self.generate_hospital_name()
            if name in visited or self.address is None:
                continue

            visited.add(name)
            res.append({
                'hospital_id': i,
                'hospital_name': name,
                'hospital_address': self.address
            })

            i += 1
        
        return pd.DataFrame(res)
```

```python
# CREATE TABLE Departments (
#     department_id INT PRIMARY KEY AUTO_INCREMENT,
#     department_name VARCHAR(45) UNIQUE,
#     hospital_id INT,

#     FOREIGN KEY (hospital_id) REFERENCES Hospitals(hospital_id)
# );

class DepartmentGenerator:
    def __init__(self, departments, df_hospital_path = 'hospitals.csv', seed = 0):
        self.departments = departments

        file_path = Path(df_hospital_path)
        if file_path.exists():
            self.df_hospital = pd.read_csv(file_path)
        elif Path('hospitals.csv').exists():
            self.df_hospital = pd.read_csv('hospitals.csv')
        else:
            raise ValueError('Hospitals table not exists.')

        self.n_hospital = self.df_hospital.shape[0]
        self.specialties = [
            'Children\'s',
            'Women\'s',
            'Heart',
            'Cancer',
            'Orthopedic',
            'Eye',
            'Psychiatric',
            'Rehabilitation'
            ]
        self.specialties_map_department = {
            'Children\'s': 'Pediatrics',
            'Women\'s': 'Obstetrics and Gynecology',
            'Heart': "Cardiology",
            'Cancer': "Oncology",
            'Orthopedic': "Orthopedics",
            'Eye': "Ophthalmology",
            'Psychiatric': "Psychiatry",
            'Rehabilitation': "Physical Medicine and Rehabilitation"
        }
        random.seed(seed)

    
    def generate(self):
        res = []
        id = 1
        for i in range(self.n_hospital):
            hospital_name = self.df_hospital.iloc[i, 2]
            department_list = set()
            # if specialty hospital
            for specialty in self.specialties:
                if specialty in hospital_name:
                    department_list.add(self.specialties_map_department[specialty])
            
            # randomly choose m departments
            m = random.randint(1, len(self.departments))
            temp = random.sample(self.departments, m)
            department_list.update(temp)

            for department in department_list:
                res.append({
                    'department_id': id,
                    'department_name': department,
                    'hospital_id': i + 1
                })
        
        return pd.DataFrame(res)
```

```python
# CREATE TABLE Doctors (
#     doctor_id INT PRIMARY KEY AUTO_INCREMENT,
#     first_name VARCHAR(45) NOT NULL,
#     last_name VARCHAR(45) NOT NULL,
#     specialization VARCHAR(45) NOT NULL,
#     phone_number VARCHAR(45) NOT NULL,
#     years_experience INT ZERO FILL,
#     hospital_id INT NOT NULL,
#     email VARCHAR(45) DEFAULT NULL,
#     department_id INT NOT NULL,

#     FOREIGN KEY (department_id) REFERENCES Departments(department_id),
#     FOREIGN KEY (hospital_id) REFERENCES Hospitals(hospital_id)
# );

class DoctorGenerator:
    def __init__(self, specialties, df_department_path = 'departments.csv', df_hospital_path = 'hospitals.csv', seed=0):
        self.fake = Faker()
        Faker.seed(seed)
        random.seed(seed)

        self.specialties = specialties

        file_path = Path(df_department_path)
        if file_path.exists():
            self.df_department = pd.read_csv(file_path)
        elif Path('departments.csv').exists():
            self.df_department = pd.read_csv('departments.csv')
        else:
            raise ValueError('Departments table not exists.')

        file_path = Path(df_hospital_path)
        if file_path.exists():
            self.df_hospitals = pd.read_csv(file_path)
        elif Path('hospitals.csv').exists():
            self.df_hospitals = pd.read_csv('hospitals.csv')
        else:
            raise ValueError('Hospitals table not exists.')


    def generate_email(self, first_name, last_name, hospital_id):
        """generate email in standardized format for each hospital"""

        domain = self.df_hospitals[self.df_hospitals['hospital_id'] == hospital_id]['hospital_name'].values
        domain = domain[0].split(' ')
        
        abbreviation = ''.join([word[0] for word in domain])

        email = f"{first_name.lower()}.{last_name.lower()}@{abbreviation}.com"

        return email

    def generate_doctors_for_hospital(self, hospital_id, n_doctors=50):
        """generate doctor for all departments"""
        hospital_departments = self.df_department[
            self.df_department['hospital_id'] == hospital_id
        ]

        if len(hospital_departments) == 0:
            return []

        if n_doctors < len(hospital_departments):
            print(f'n_doctors is too small for hospital {hospital_id}, increase n_doctors to 2 * number of departments.')
            n_doctors = 2 * len(hospital_departments)

        doctors = []

        # 1. make sure that all departments are assigned at least one doctor
        for _, dept_row in hospital_departments.iterrows():
            department_name = dept_row['department_name']
            department_id = dept_row['department_id']

            # randomly select a specialty
            if department_name in self.specialties:
                specialization = random.choice(self.specialties[department_name])
            else:
                specialization = department_name  # directly use department name

            first_name = self.fake.first_name()
            last_name = self.fake.last_name()

            doctor = {
                'first_name': first_name,
                'last_name': last_name,
                'specialization': specialization,
                'phone_number': self.fake.phone_number(),
                'years_experience': random.randint(1, 35),
                'hospital_id': hospital_id,
                'email': self.generate_email(first_name, last_name, hospital_id),
                'department_id': department_id
            }
            doctors.append(doctor)

        # 2. randomly assign the rest of the doctors to departments excluding the doctors above
        remaining_doctors = n_doctors - len(hospital_departments)

        for _ in range(remaining_doctors):
            # randomly select a department
            dept_row = hospital_departments.sample(n=1).iloc[0]
            department_name = dept_row['department_name']
            department_id = dept_row['department_id']

            # randomly select a specialty
            if department_name in self.specialties:
                specialization = random.choice(self.specialties[department_name])
            else:
                specialization = department_name

            first_name = self.fake.first_name()
            last_name = self.fake.last_name()

            doctor = {
                'first_name': first_name,
                'last_name': last_name,
                'specialization': specialization,
                'phone_number': self.fake.phone_number(),
                'years_experience': random.randint(1, 35),
                'hospital_id': hospital_id,
                'email': self.generate_email(first_name, last_name, hospital_id),
                'department_id': department_id
            }
            doctors.append(doctor)

        return doctors

    def generate(self, n_doctors_per_hospital=50):
        """generate doctors for all hospitals"""
        all_doctors = []
        doctor_id = 1

        # get all hospital_id
        hospital_ids = self.df_department['hospital_id'].unique()

        for hospital_id in hospital_ids:
            doctors = self.generate_doctors_for_hospital(hospital_id, n_doctors_per_hospital)

            # add doctor_id
            for doctor in doctors:
                doctor['doctor_id'] = doctor_id
                all_doctors.append(doctor)
                doctor_id += 1

        return pd.DataFrame(all_doctors)
```

```python
# CREATE TABLE Patients (
#     patient_id INT PRIMARY KEY AUTO_INCREMENT COMMENT 'Unique ID for each patient',
#     first_name VARCHAR(45) NOT NULL COMMENT 'Patient\'s first name',
#     last_name VARCHAR(45) NOT NULL COMMENT 'Patient\'s last name',
#     gender ENUM('M', 'F') DEFAULT NULL COMMENT 'Gender (M/F)',
#     date_of_birth DATE NOT NULL COMMENT 'Date of birth (YYYY-MM-DD)',
#     contact_number VARCHAR(20) DEFAULT NULL COMMENT 'Phone number',
#     address VARCHAR(100) COMMENT 'Address',
#     registration_date DATE DEFAULT CURRENT_DATE COMMENT 'Date of first registration at the hospital',
#     insurance_provider VARCHAR(45) DEFAULT NULL COMMENT 'Insurance company name',
#     insurance_number VARCHAR(45) DEFAULT NULL,
#     email VARCHAR(45),
#     ssn VARCHAR(45)
# );

import datetime
from datetime import timedelta

class PatientGenerator:
    def __init__(self, seed=0):
        self.fake = Faker()
        Faker.seed(seed)
        random.seed(seed)
        
        # insurance providers
        self.insurance_providers = [
            'Blue Cross Blue Shield',
            'UnitedHealthcare',
            'Aetna',
            'Cigna',
            'Humana',
            'Kaiser Permanente',
            'Anthem',
            'Medicare',
            'Medicaid',
            None
        ]
    
    def generate_date_of_birth(self):
        """
        generate date of birth (age between 0 and 100)
        """
        today = datetime.date.today()
        age = random.randint(0, 100)
        birth_year = today.year - age
        start_date = datetime.date(birth_year, 1, 1)
        end_date = datetime.date(birth_year, 12, 31)
        
        if end_date > today:
            end_date = today
        
        return self.fake.date_between(start_date=start_date, end_date=end_date)
    
    def generate_registration_date(self, date_of_birth):
        """
        generate registration date with 80% of patients registered within past 5 years
        """
        today = datetime.date.today()
        
        if random.random() < 0.8:
            earliest_date = max(date_of_birth, today - timedelta(days=5*365))
        else:
            earliest_date = date_of_birth
        
        return self.fake.date_between(start_date=earliest_date, end_date=today)
    
    
    def generate_insurance_number(self, insurance_provider):
        """
        generate insurance number if insurance provider exists else return None
        """
        if insurance_provider is None:
            return None
        
        # format: letters * 3 + numbers * 9
        prefix = ''.join(random.choices('ABCDEFGHIJKLMNOPQRSTUVWXYZ', k=3))
        number = ''.join(random.choices('0123456789', k=9))
        return f"{prefix}{number}"
    
    def generate_email(self, first_name, last_name):
        """
        generate emails for 70% patients
        """
        if random.random() < 0.3:
            return None
        
        domains = ['gmail.com', 'yahoo.com', 'hotmail.com', 'outlook.com', 'icloud.com']
        # add random numbers 
        random_num = random.randint(1, 999)
        email = f"{first_name.lower()}.{last_name.lower()}{random_num}@{random.choice(domains)}"
        return email
    
    
    def generate(self, n):
        """
        generate n patient data
        """
        patients = []
        
        for i in range(n):
            first_name = self.fake.first_name()
            last_name = self.fake.last_name()
            gender = random.choice(['M', 'F'])
            address = self.fake.address()
            address = address.replace('\n', ', ').strip()
            
            date_of_birth = self.generate_date_of_birth()
            registration_date = self.generate_registration_date(date_of_birth)
            
            insurance_provider = random.choice(self.insurance_providers)
            insurance_number = self.generate_insurance_number(insurance_provider)
            
            patient = {
                'patient_id': i + 1,
                'first_name': first_name,
                'last_name': last_name,
                'gender': gender,
                'date_of_birth': date_of_birth,
                'contact_number': self.fake.phone_number(),
                'address': address,
                'registration_date': registration_date,
                'insurance_provider': insurance_provider,
                'insurance_number': insurance_number,
                'email': self.generate_email(first_name, last_name),
                'ssn': self.fake.ssn()
            }
            
            patients.append(patient)


        df = pd.DataFrame(patients)
        
        column_order = [
            'patient_id', 'first_name', 'last_name', 'gender', 
            'date_of_birth', 'contact_number', 'address', 'registration_date',
            'insurance_provider', 'insurance_number', 'email', 'ssn'
        ]
        
        return df[column_order]
```

```python
# CREATE TABLE Appointments (
#     appointment_id INT PRIMARY KEY AUTO_INCREMENT,
#     patient_id INT NOT NULL,
#     doctor_id INT NOT NULL,
#     reason_for_visit VARCHAR(45) NOT NULL,
#     appointment_type ENUM('Regular', 'Diagnostic', 'Treatment', 'Others') COMMENT 'Status (Regular, Diagnostic, Treatment, Others)',
#     status ENUM('Scheduled', 'Completed', 'Cancelled') COMMENT 'Scheduled, Completed, Cancelled',
#     appointment_time TIME NOT NULL COMMENT 'Time of the appointment',
#     appointment_date DATE NOT NULL COMMENT 'Date of the appointment',

#     FOREIGN KEY (patient_id) REFERENCES Patients(patient_id),
#     FOREIGN KEY (doctor_id) REFERENCES Doctors(doctor_id)
# );

import datetime

class AppointmentGenerator:
    def __init__(self, reasons_for_visit, reason_to_departments, reason_to_appointment_type, df_doctor_path = 'doctors.csv', df_department_path = 'departments.csv', 
    df_patient_path = 'patients.csv', start_date = None, end_date = datetime.date.today(), seed = 0):
        self.fake = Faker()
        Faker.seed(seed)
        random.seed(seed)

        if start_date is None:
            self.start_date = end_date - datetime.timedelta(days=5*365)  # record appointments for past 5 years
        else:
            self.start_date = start_date
        self.end_date = end_date + datetime.timedelta(days=365)  # accept appointments for 1 year in the future

        self.reasons_for_visit = reasons_for_visit
        self.reason_to_departments = reason_to_departments
        self.reason_to_appointment_type = reason_to_appointment_type
        self.status = ['Scheduled', 'Completed', 'Cancelled']

        file_path_dept = Path(df_department_path)
        if file_path_dept.exists():
            self.df_departments = pd.read_csv(file_path_dept)
        elif Path('departments.csv').exists():
            self.df_departments = pd.read_csv('departments.csv')
        else:
            raise ValueError('Departments table not exists.')

        file_path_doc = Path(df_doctor_path)
        if file_path_doc.exists():
            self.df_doctors = pd.read_csv(file_path_doc)
        elif Path('doctors.csv').exists():
            self.df_doctors = pd.read_csv('doctors.csv')
        else:
            raise ValueError('Doctors table not exists.')

        file_path_pat = Path(df_patient_path)
        if file_path_pat.exists():
            self.df_patients = pd.read_csv(file_path_pat)
        elif Path('patients.csv').exists():
            self.df_patients = pd.read_csv('patients.csv')
        else:
            raise ValueError('Patients table not exists.')
        

    def generate_appointment(self):
        # select patient
        patient = self.df_patients.sample(n=1).iloc[0]
        
        # select reason for visit
        reason = random.choice(self.reasons_for_visit)
        
        # select appointment type
        appointment_type = self.reason_to_appointment_type[reason]

        # choose department according to reason
        available_departments = self.reason_to_departments[reason]
        department = random.choice(available_departments)

        # èŽ·å–æä¾›è¯¥ç§‘å®¤çš„åŒ»é™¢éƒ¨é—¨è®°å½•
        department_records = self.df_departments[self.df_departments['department_name'] == department]
        
        if department_records.empty:
            return None
            
        # éšæœºé€‰æ‹©ä¸€ä¸ªéƒ¨é—¨è®°å½•
        selected_dept = department_records.sample(n=1).iloc[0]
        department_id = selected_dept['department_id']
        hospital_id = selected_dept['hospital_id']
        
        available_doctors = self.df_doctors[
            self.df_doctors['department_id'] == department_id
        ]
        
        if len(available_doctors) == 0:
            return None
        
        # randomly select one doctor
        doctor = available_doctors.sample(n=1).iloc[0]
        
        # generate appointment date and time
        patient_reg_date = pd.to_datetime(patient['registration_date']).date()
        appointment_start = max(patient_reg_date, self.start_date)  # appointment date should be after patients' registration date
        
        appointment_date = self.fake.date_between(
            start_date=appointment_start,
            end_date=self.end_date
        )
        
        # reasonable appointment time
        hour = random.randint(8, 16)  # 8 AM to 4 PM
        minute = random.choice([0, 15, 30, 45])  # 15 minutes interval
        appointment_time = datetime.time(hour, minute)

        # generate appointment status
        # for future appointments, status can only be Scheduled
        # for past appointments, status can be Completed or Cancelled

        if appointment_date > datetime.date.today():
            status = 'Scheduled'
        else:
            status = random.choice(['Completed', 'Cancelled'])
            if appointment_type in ['Treatment', 'Others']:  # for treatment, lower the probability of cancelling appointment
                status = 'Cancelled' if random.random() < 0.1 else 'Completed'

        return {
            'patient_id': patient['patient_id'],
            'doctor_id': doctor['doctor_id'],
            'reason_for_visit': reason,
            'appointment_type': appointment_type,
            'status': status,
            'appointment_time': appointment_time,
            'appointment_date': appointment_date,
        }
    
    def generate(self, n):
        res = []
        for i in range(n):
            temp = self.generate_appointment()
            if temp is None:
                continue
            temp['appointment_id'] = i + 1
            res.append(temp)

        df = pd.DataFrame(res)
        columns = ['appointment_id', 'patient_id', 'doctor_id', 'reason_for_visit', 'appointment_type', 'status', 'appointment_time', 'appointment_date']

        return df[columns]
```

```python
# CREATE TABLE Treatments (
#     treatment_id INT PRIMARY KEY AUTO_INCREMENT,
#     appointment_id INT NOT NULL,
#     treatment_type VARCHAR(45) NOT NULL COMMENT 'Type of treatment (e.g., MRI, X-ray)',
#     description TEXT COMMENT 'Notes or procedure details',
#     cost DECIMAL(10, 2) NOT NULL COMMENT 'Cost of treatment',
#     treatment_date DATE DEFAULT CURRENT_DATE COMMENT 'Date when treatment was given',

#     FOREIGN KEY (appointment_id) REFERENCES Appointments(appointment_id)
# );

import datetime
from pathlib import Path

class TreatmentGenerator:
    def __init__(self, reason_to_treatment_types, treatment_cost_ranges, treatment_type_to_description_level,
     df_appointments_path = 'appointments.csv', seed = 0):
        self.faker = Faker()
        Faker.seed(seed)
        random.seed(seed)
        self.reason_to_treatment_types = reason_to_treatment_types
        self.treatment_cost_ranges = treatment_cost_ranges
        self.treatment_type_to_description_level = treatment_type_to_description_level

        file_path = Path(df_appointments_path)
        if file_path.exists():
            self.df_appointments = pd.read_csv(file_path)
        elif Path('appointments.csv').exists():
            self.df_appointments = pd.read_csv('appointments.csv')
        else:
            raise ValueError('Appointments table not exists.')

    def generate(self, end_date = datetime.date.today()):
        """
        generate treatment for every past completed appointment
        """
        res = []
        treatment_id = 1 
        for _, appointment in self.df_appointments.iterrows():
            if appointment['status'] == 'Completed':

                # get appointment id
                appointment_id = appointment['appointment_id']

                # get appointment date
                start_date = pd.Timestamp(appointment['appointment_date']).date()  # first transfer to pd.timestamp, then to pydatetime

                # get reason for visit from appointments
                reason = appointment['reason_for_visit']

                # get all possible treatments
                available_treatments = self.reason_to_treatment_types.get(reason)

                # randomly select treatments
                treatment_types = random.sample(available_treatments, k = min(random.randint(1, 3), len(available_treatments)))

                # get cost
                total_cost = 0
                for treatment in treatment_types:
                    cost_range = self.treatment_cost_ranges.get(treatment, (100, 1000))
                    total_cost += random.uniform(cost_range[0], cost_range[1])
                total_cost = round(total_cost, 2)

                # get descroption
                descriptions = [treatment_type_to_description_level[treatment] for treatment in treatment_types]
                if 'Advanced protocol' in descriptions:
                    description = 'Advanced protocol'
                elif 'Standard procedure' in descriptions:
                    description = 'Standard procedure'
                else:
                    description = 'Basic screening'

                # get treatment date
                if random.random() < 0.1:
                    # treatment may be 1-7 days later
                    days_after = random.randint(1, 7)
                    treatment_date = start_date + datetime.timedelta(days=days_after)
                    # make sure it's not later than end_date(today)
                    if treatment_date > end_date:
                        treatment_date = end_date
                else:
                    treatment_date = start_date

                res.append({
                    'treatment_id': treatment_id,
                    'appointment_id': appointment_id,
                    'treatment_types': ', '.join(treatment_types),
                    'description': description,
                    'cost': total_cost,
                    'treatment_date': treatment_date,
                })
                
                treatment_id += 1
        
        df = pd.DataFrame(res)

        return df
```

```python
# CREATE TABLE DoctorNotes (
#     note_id INT PRIMARY KEY AUTO_INCREMENT,
#     patient_id INT NOT NULL,
#     appointment_id INT NOT NULL,
#     doctor_id INT NOT NULL,
#     note_date DATETIME NOT NULL DEFAULT CURRENT_TIMESTAMP,
    
#     diagnosis TEXT COMMENT 'Diagnosis for this visit',
#     symptoms TEXT COMMENT 'Symptoms reported/observed',
#     next_steps TEXT COMMENT 'Recommended next steps',
#     prescription TEXT COMMENT 'Medications prescribed',
#     doctor_comments TEXT COMMENT 'Additional notes from doctor',
#     follow_up_needed ENUM('Yes', 'No') DEFAULT 'No',
#     follow_up_date DATE COMMENT 'Suggested follow-up date',
    
#     FOREIGN KEY (patient_id) REFERENCES Patients(patient_id),
#     FOREIGN KEY (appointment_id) REFERENCES Appointments(appointment_id),
#     FOREIGN KEY (doctor_id) REFERENCES Doctors(doctor_id)
# );


class DoctorNotesIntegrator:
    def __init__(self, kaggle_specialty_to_department, kaggle_csv_path, 
                 df_appointments_path='appointments.csv',
                 df_doctors_path='doctors.csv',
                 df_departments_path='departments.csv',
                 seed=0):
        """
        integrate reports.csv to the DoctorNotes table
        """
        random.seed(seed)

        self.df_kaggle = pd.read_csv(kaggle_csv_path)
        self.df_appointments = pd.read_csv(df_appointments_path)
        self.df_doctors = pd.read_csv(df_doctors_path)
        self.df_departments = pd.read_csv(df_departments_path)

        # for completed appointments
        self.completed_appointments = self.df_appointments[
            self.df_appointments['status'] == 'Completed'
        ].copy()

        self.specialty_mapping = kaggle_specialty_to_department

    def parse_medical_report(self, report_text, medical_specialty):
        result = {
            'diagnosis': '',
            'symptoms': '',
            'next_steps': '',
            'prescription': '',
            'doctor_comments': '',
            'follow_up_needed': 'No',
        }

        report = str(report_text).strip()
        report_lower = report.lower()

        # segmentation 
        sections = {}

        # common segment keywords
        section_keywords = {
            'symptoms': ['chief complaint:', 'subjective:', 'history:', 'presenting complaint:', 'reason for visit:'],
            'diagnosis': ['diagnosis:', 'assessment:', 'impression:', 'findings:', 'clinical impression:', 'final diagnosis:'],
            'next_steps': ['plan:', 'recommendations:', 'treatment plan:', 'management:', 'disposition:'],
            'prescription': ['medications:', 'prescription:', 'rx:', 'drugs prescribed:'],
            'comments': ['note:', 'comments:', 'additional notes:', 'remarks:']
        }

        # extract symptoms
        for keyword in section_keywords['symptoms']:
            if keyword in report_lower:
                start = report_lower.index(keyword) + len(keyword)
                # store the start position of the next part
                end = len(report)
                for all_keywords in section_keywords.values():
                    for next_kw in all_keywords:
                        if next_kw in report_lower[start:] and next_kw != keyword:
                            temp_end = start + report_lower[start:].index(next_kw)
                            if temp_end < end:
                                end = temp_end
                result['symptoms'] = report[start:end].strip()[:1000]
                break

        # extract diagnosis
        for keyword in section_keywords['diagnosis']:
            if keyword in report_lower:
                start = report_lower.index(keyword) + len(keyword)
                end = len(report)
                for all_keywords in section_keywords.values():
                    for next_kw in all_keywords:
                        if next_kw in report_lower[start:] and next_kw != keyword:
                            temp_end = start + report_lower[start:].index(next_kw)
                            if temp_end < end:
                                end = temp_end
                result['diagnosis'] = report[start:end].strip()[:1000]
                break

        # extract next_step
        for keyword in section_keywords['next_steps']:
            if keyword in report_lower:
                start = report_lower.index(keyword) + len(keyword)
                end = len(report)
                for all_keywords in section_keywords.values():
                    for next_kw in all_keywords:
                        if next_kw in report_lower[start:] and next_kw != keyword:
                            temp_end = start + report_lower[start:].index(next_kw)
                            if temp_end < end:
                                end = temp_end
                result['next_steps'] = report[start:end].strip()[:1000]
                break

        # extract prescription
        for keyword in section_keywords['prescription']:
            if keyword in report_lower:
                start = report_lower.index(keyword) + len(keyword)
                end = len(report)
                result['prescription'] = report[start:end].strip()[:1000]
                break

        # take the whole report as doctor_comments
        if not any([result['symptoms'], result['diagnosis'], result['next_steps']]):
            result['doctor_comments'] = report[:2000]
        else:
            # otherwise take the rest part as doctor_comments
            result['doctor_comments'] = report[:1000]

        # find follow up keywords
        followup_keywords = ['follow-up', 'follow up', 'return', 'recheck', 'see again in']
        if any(keyword in report_lower for keyword in followup_keywords):
            result['follow_up_needed'] = 'Yes'

        return result

    def get_matching_doctor(self, appointment_row, kaggle_specialty):

        # get possible departments
        possible_departments = self.specialty_mapping.get(kaggle_specialty, [])

        if not possible_departments:
            # if no available doctors, return None
            return 

        # get doctor ids from departments table
        doctor_id = appointment_row['doctor_id']
        doctor_row = self.df_doctors[self.df_doctors['doctor_id'] == doctor_id]

        if len(doctor_row) > 0:
            doctor_dept_id = doctor_row.iloc[0]['department_id']
            doctor_dept = self.df_departments[
                self.df_departments['department_id'] == doctor_dept_id
            ]

            if len(doctor_dept) > 0:
                doctor_dept_name = doctor_dept.iloc[0]['department_name']

                # if doctor's department matches specialty in dataset, directly use
                if doctor_dept_name in possible_departments:
                    return doctor_id

        # otherwise, randomly select a doctor from targeted department
        matching_dept_ids = self.df_departments[
            self.df_departments['department_name'].isin(possible_departments)
        ]['department_id'].tolist()

        if matching_dept_ids:
            matching_doctors = self.df_doctors[
                self.df_doctors['department_id'].isin(matching_dept_ids)
            ]

            if len(matching_doctors) > 0:
                return matching_doctors.sample(n=1).iloc[0]['doctor_id']

        # if not found, return None
        return None

    
    def generate_integrated_notes(self):    
        """
        Traverse all Kaggle reports and find the corresponding appointment for each report.
        """
        results = []
        note_id = 1
        used_appointments = set()  # track used appointment
        skipped_reports = 0
        
        print(f"Total reports to process: {len(self.df_kaggle)}")
        print(f"Total completed appointments: {len(self.completed_appointments)}")
        
        # traverse every report
        for idx, kaggle_row in self.df_kaggle.iterrows():
            medical_specialty = kaggle_row['medical_specialty']
            report_text = kaggle_row['report']
            
            # get corresponding department id
            possible_departments = self.specialty_mapping.get(medical_specialty, [])
            
            if not possible_departments:
                skipped_reports += 1
                if skipped_reports <= 5:  # print warning for first 5 skipped reports
                    print(f"Warning: No department mapping for specialty '{medical_specialty}'")
                continue
            
            # get department_ids
            matching_dept_ids = self.df_departments[
                self.df_departments['department_name'].isin(possible_departments)
            ]['department_id'].tolist()
            
            if not matching_dept_ids:
                skipped_reports += 1
                continue
            
            # find available doctors
            matching_doctors = self.df_doctors[
                self.df_doctors['department_id'].isin(matching_dept_ids)
            ]
            
            if len(matching_doctors) == 0:
                skipped_reports += 1
                continue
            
            # get doctor_ids
            matching_doctor_ids = matching_doctors['doctor_id'].tolist()
            
            # get these doctors' completed appointments
            available_appointments = self.completed_appointments[
                (self.completed_appointments['doctor_id'].isin(matching_doctor_ids)) &
                (~self.completed_appointments['appointment_id'].isin(used_appointments))
            ]
            
            if len(available_appointments) == 0:
                # if no available appointment, skip
                skipped_reports += 1
                continue
            
            # randomly select one appointment
            selected_appointment = available_appointments.sample(n=1).iloc[0]
            
            used_appointments.add(selected_appointment['appointment_id'])
            
            # parse the report
            parsed_report = self.parse_medical_report(report_text, medical_specialty)
            
            appointment_id = selected_appointment['appointment_id']
            patient_id = selected_appointment['patient_id']
            doctor_id = selected_appointment['doctor_id']
            
            # note_date = appointment date + random time
            appointment_date = pd.to_datetime(selected_appointment['appointment_date'])
            note_datetime = appointment_date + timedelta(hours=random.randint(0, 8))
            
            # get random follow up dates
            follow_up_date = None
            if parsed_report['follow_up_needed'] == 'Yes':
                weeks_later = random.choice([1, 2, 4, 8, 12])
                follow_up_date = (appointment_date + timedelta(weeks=weeks_later)).date()
            
            note_record = {
                'note_id': note_id,
                'patient_id': patient_id,
                'appointment_id': appointment_id,
                'doctor_id': doctor_id,
                'note_date': note_datetime,
                'diagnosis': parsed_report['diagnosis'] if parsed_report['diagnosis'] else None,
                'symptoms': parsed_report['symptoms'] if parsed_report['symptoms'] else None,
                'next_steps': parsed_report['next_steps'] if parsed_report['next_steps'] else None,
                'prescription': parsed_report['prescription'] if parsed_report['prescription'] else None,
                'doctor_comments': parsed_report['doctor_comments'] if parsed_report['doctor_comments'] else None,
                'follow_up_needed': parsed_report['follow_up_needed'],
                'follow_up_date': follow_up_date
            }
            
            results.append(note_record)
            note_id += 1
            
            # print progress
            if note_id % 100 == 0:
                print(f"Processed {note_id} reports...")
        
        print(f"\n=== Generation Complete ===")
        print(f"Total reports processed: {len(self.df_kaggle)}")
        print(f"Successfully generated notes: {len(results)}")
        print(f"Skipped reports: {skipped_reports}")
        print(f"Unique appointments used: {len(used_appointments)}")
        
        if len(results) == 0:
            print("\nWarning: No notes generated! Check your specialty mappings.")
            return pd.DataFrame()
        
        df = pd.DataFrame(results)
        
        column_order = [
            'note_id', 'patient_id', 'appointment_id', 'doctor_id', 'note_date',
            'diagnosis', 'symptoms', 'next_steps', 'prescription', 
            'doctor_comments', 'follow_up_needed', 'follow_up_date'
        ]
        
        return df[column_order]
```

```python
# CREATE TABLE DoctorNotes (
#     note_id INT PRIMARY KEY AUTO_INCREMENT,
#     patient_id INT NOT NULL,
#     appointment_id INT NOT NULL,
#     doctor_id INT NOT NULL,
#     note_date DATETIME NOT NULL DEFAULT CURRENT_TIMESTAMP,
    
#     diagnosis TEXT COMMENT 'Diagnosis for this visit',
#     symptoms TEXT COMMENT 'Symptoms reported/observed',
#     next_steps TEXT COMMENT 'Recommended next steps',
#     prescription TEXT COMMENT 'Medications prescribed',
#     doctor_comments TEXT COMMENT 'Additional notes from doctor',
#     follow_up_needed ENUM('Yes', 'No') DEFAULT 'No',
#     follow_up_date DATE COMMENT 'Suggested follow-up date',
    
#     FOREIGN KEY (patient_id) REFERENCES Patients(patient_id),
#     FOREIGN KEY (appointment_id) REFERENCES Appointments(appointment_id),
#     FOREIGN KEY (doctor_id) REFERENCES Doctors(doctor_id)
# );


class DoctorNotesIntegrator:
    def __init__(self, kaggle_specialty_to_department, kaggle_csv_path, 
                 df_appointments_path='appointments.csv',
                 df_doctors_path='doctors.csv',
                 df_departments_path='departments.csv',
                 seed=0):
        """
        integrate reports.csv to the DoctorNotes table
        """
        random.seed(seed)

        self.df_kaggle = pd.read_csv(kaggle_csv_path)
        self.df_appointments = pd.read_csv(df_appointments_path)
        self.df_doctors = pd.read_csv(df_doctors_path)
        self.df_departments = pd.read_csv(df_departments_path)

        # for completed appointments
        self.completed_appointments = self.df_appointments[
            self.df_appointments['status'] == 'Completed'
        ].copy()

        self.specialty_mapping = kaggle_specialty_to_department

    def parse_medical_report(self, report_text, medical_specialty):
        result = {
            'diagnosis': '',
            'symptoms': '',
            'next_steps': '',
            'prescription': '',
            'doctor_comments': '',
            'follow_up_needed': 'No',
        }

        report = str(report_text).strip()
        report_lower = report.lower()

        # segmentation 
        sections = {}


        # take the whole report as doctor_comments
        result['doctor_comments'] = report[:]

        # find follow up keywords
        followup_keywords = ['follow-up', 'follow up', 'return', 'recheck', 'see again in']
        if any(keyword in report_lower for keyword in followup_keywords):
            result['follow_up_needed'] = 'Yes'

        return result

    def get_matching_doctor(self, appointment_row, kaggle_specialty):

        # get possible departments
        possible_departments = self.specialty_mapping.get(kaggle_specialty, [])

        if not possible_departments:
            # if no available doctors, return None
            return 

        # get doctor ids from departments table
        doctor_id = appointment_row['doctor_id']
        doctor_row = self.df_doctors[self.df_doctors['doctor_id'] == doctor_id]

        if len(doctor_row) > 0:
            doctor_dept_id = doctor_row.iloc[0]['department_id']
            doctor_dept = self.df_departments[
                self.df_departments['department_id'] == doctor_dept_id
            ]

            if len(doctor_dept) > 0:
                doctor_dept_name = doctor_dept.iloc[0]['department_name']

                # if doctor's department matches specialty in dataset, directly use
                if doctor_dept_name in possible_departments:
                    return doctor_id

        # otherwise, randomly select a doctor from targeted department
        matching_dept_ids = self.df_departments[
            self.df_departments['department_name'].isin(possible_departments)
        ]['department_id'].tolist()

        if matching_dept_ids:
            matching_doctors = self.df_doctors[
                self.df_doctors['department_id'].isin(matching_dept_ids)
            ]

            if len(matching_doctors) > 0:
                return matching_doctors.sample(n=1).iloc[0]['doctor_id']

        # if not found, return None
        return None

    
    def generate_integrated_notes(self):    
        """
        Traverse all Kaggle reports and find the corresponding appointment for each report.
        """
        results = []
        note_id = 1
        used_appointments = set()  # track used appointment
        skipped_reports = 0
        
        print(f"Total reports to process: {len(self.df_kaggle)}")
        print(f"Total completed appointments: {len(self.completed_appointments)}")
        
        # traverse every report
        for idx, kaggle_row in self.df_kaggle.iterrows():
            medical_specialty = kaggle_row['medical_specialty']
            report_text = kaggle_row['report']
            
            # get corresponding department id
            possible_departments = self.specialty_mapping.get(medical_specialty, [])
            
            if not possible_departments:
                skipped_reports += 1
                if skipped_reports <= 5:  # print warning for first 5 skipped reports
                    print(f"Warning: No department mapping for specialty '{medical_specialty}'")
                continue
            
            # get department_ids
            matching_dept_ids = self.df_departments[
                self.df_departments['department_name'].isin(possible_departments)
            ]['department_id'].tolist()
            
            if not matching_dept_ids:
                skipped_reports += 1
                continue
            
            # find available doctors
            matching_doctors = self.df_doctors[
                self.df_doctors['department_id'].isin(matching_dept_ids)
            ]
            
            if len(matching_doctors) == 0:
                skipped_reports += 1
                continue
            
            # get doctor_ids
            matching_doctor_ids = matching_doctors['doctor_id'].tolist()
            
            # get these doctors' completed appointments
            available_appointments = self.completed_appointments[
                (self.completed_appointments['doctor_id'].isin(matching_doctor_ids)) &
                (~self.completed_appointments['appointment_id'].isin(used_appointments))
            ]
            
            if len(available_appointments) == 0:
                # if no available appointment, skip
                skipped_reports += 1
                continue
            
            # randomly select one appointment
            selected_appointment = available_appointments.sample(n=1).iloc[0]
            
            used_appointments.add(selected_appointment['appointment_id'])
            
            # parse the report
            parsed_report = self.parse_medical_report(report_text, medical_specialty)
            
            appointment_id = selected_appointment['appointment_id']
            patient_id = selected_appointment['patient_id']
            doctor_id = selected_appointment['doctor_id']
            
            # note_date = appointment date + random time
            appointment_date = pd.to_datetime(selected_appointment['appointment_date'])
            note_datetime = appointment_date + timedelta(hours=random.randint(0, 8))
            
            # get random follow up dates
            follow_up_date = None
            if parsed_report['follow_up_needed'] == 'Yes':
                weeks_later = random.choice([1, 2, 4, 8, 12])
                follow_up_date = (appointment_date + timedelta(weeks=weeks_later)).date()
            
            note_record = {
                'note_id': note_id,
                'patient_id': patient_id,
                'appointment_id': appointment_id,
                'doctor_id': doctor_id,
                'note_date': note_datetime,
                'diagnosis': None,
                'symptoms': None,
                'next_steps': None,
                'prescription': None,
                'doctor_comments': parsed_report['doctor_comments'] if parsed_report['doctor_comments'] else None,
                'follow_up_needed': parsed_report['follow_up_needed'],
                'follow_up_date': follow_up_date
            }
            
            results.append(note_record)
            note_id += 1
            
            # print progress
            if note_id % 100 == 0:
                print(f"Processed {note_id} reports...")
        
        print(f"\n=== Generation Complete ===")
        print(f"Total reports processed: {len(self.df_kaggle)}")
        print(f"Successfully generated notes: {len(results)}")
        print(f"Skipped reports: {skipped_reports}")
        print(f"Unique appointments used: {len(used_appointments)}")
        
        if len(results) == 0:
            print("\nWarning: No notes generated! Check your specialty mappings.")
            return pd.DataFrame()
        
        df = pd.DataFrame(results)
        
        column_order = [
            'note_id', 'patient_id', 'appointment_id', 'doctor_id', 'note_date',
            'diagnosis', 'symptoms', 'next_steps', 'prescription', 
            'doctor_comments', 'follow_up_needed', 'follow_up_date'
        ]
        
        return df[column_order]
```

```python
integrator = DoctorNotesIntegrator(
    kaggle_specialty_to_department = kaggle_specialty_to_department,
    kaggle_csv_path='doctor_notes/reports.csv',
    df_appointments_path='appointments.csv',
    df_doctors_path='doctors.csv',
    df_departments_path='departments.csv',
    seed=42
)


df_doctor_notes = integrator.generate_integrated_notes()

# follow up stats
print(f"\nFollow-up distribution:")
print(df_doctor_notes['follow_up_needed'].value_counts())

print(f"\nFields with data:")
for col in ['diagnosis', 'symptoms', 'next_steps', 'prescription']:
    count = df_doctor_notes[col].notna().sum()
    print(f"  {col}: {count} ({count/len(df_doctor_notes)*100:.1f}%)")
```
